# Stack and __queue

## 1. Stack  

A **Stack** is a linear data structure that follows the **LIFO (Last In, First Out)** principle. It holds a specific type of data, such as:  
- `int`
- `float`
- `char`
- Custom data structures  
- Pairs of data  

### Key Operations in Stack:
A stack supports the following four fundamental operations:

```plaintext
1. Push()  - Inserts an element at the top of the stack.
2. Pop()   - Removes the top element from the stack.
3. Top()   - Retrieves the top element without removing it.
4. __size()  - Returns the number of elements in the stack.



# Let's Implement the Stack Using Array

````python
top = "cah"
if type(top) == str:
    print(top)
````

````python
class Stack:
    def __init__(self, size):
        self.__stack = [None]*size
        self.__size = size
        self.__top = -1

    def push(self, value):
        if self.__top == self.__size-1:
            print("Stack is OverFlow")
            return
        else:
            self.__top += 1
            self.__stack[self.__top] = value

    def pop(self):
        if self.__top == -1:
            return "Stack is Empty !"
        else:
            popped_val = self.__stack[self.__top]
            self.__stack[self.__top] = "None"
            self.__top-=1
            return f" Popped Value {popped_val}"

    def peek(self):
        # return self.stack[self.top]
        if self.__top == -1:
            return "Stack is Empty !"
        else:
            return self.stack[self.top]

    def is_empty(self):
        return self.__top == -1

    def size(self):
        return self.__top+1
    
    def display(self):
        if self.__top == -1:
            return "Stack is Empty"
        else:
            return self.__stack[:self.__top+1]

st = Stack(4)
st.push(1)
st.push(2)
st.push(3)
st.push(4)

st.display()
````

## 2. __queue  

A **______________queue** is a linear data structure that follows the **FIFO (First In, First Out)** principle. It holds a specific type of data, such as:  
- `int`
- `float`
- `char`
- Custom data structures  
- Pairs of data  

### Key Operations in __queue:

```plaintext

3. Front()   - Retrieves the front element without removing it.
